```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.metrics import adjusted_rand_score

#Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data

#Step 2: Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Visualize dendrogram to understand the hierarchical structure
plt.figure(figsize=(10, 7))
plt.title("Iris Dendrogram")
dend = shc.dendrogram(shc.linkage(X_scaled, method='ward'))
plt.show()

# Step 4: Apply Agglomerative Hierarchical Clustering
agg_cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_pred = agg_cluster.fit_predict(X_scaled)

# Step 5: Visualize the clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap='rainbow')
plt.title('Agglomerative Clustering - Iris Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Optional: Compare with actual labels from the dataset
print("Actual labels:", iris.target)
print("Cluster labels:", y_pred)

accuracy = adjusted_rand_score(iris.target, y_pred)
print("Adjusted Rand Index (Accuracy):", accuracy)
```
